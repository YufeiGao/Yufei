```python
# find missing data column & percentage
empty = []
for key, value in loan_data.iteritems():
    empty_number = 0
    for i in value:
        try:
            if pandas.isnull(i): empty_number += 1
        except TypeError:
            None
    if empty_number != 0:
        empty.append((key, float(empty_number) / float(len(value))))
empty
```

```python
# columns from A to Z:
# 'id','member_id','emp_title','url','dti' could be dropped
# loan_data.drop(key, 1, inplace= True)
# empty cell in column 'dti' could be replace by average number
DTI = loan_data['dti']
non_empty_number = 0
summation = 0.0
for element in DTI:
    if not pandas.isnull(element): 
        non_empty_number += 1
        summation += element
avg = summation / non_empty_number
for i in xrange(len(DTI)):
    if pandas.isnull(DTI[i]): DTI[i] = avg
```
